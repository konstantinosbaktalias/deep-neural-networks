```python
import numpy as np
import matplotlib.pyplot as plt
import mnist, os, sys

path = os.getcwd()
sys.path.append(path + '../../../')

from neuralnetwork import NeuralNetwork
```

```python
train_data = mnist.train_images().reshape(60000, 784)
train_labels = mnist.train_labels()
```

```python
def transform(arr):
    temp = np.zeros((len(arr), 10)) 
    for i in range(len(arr)):
        temp[i][arr[i] - 1] = 1.0
    return temp
    
train_labels = transform(train_labels)
train_data = (train_data / 255) 
```

```python
nn = NeuralNetwork([784, 64, 64, 10])
# nn.load('model')
nn.back_prop(
    X=train_data,
    Y=train_labels,
    epochs=10,
    rate=0.1,
    plot=True
    )
```

```python
test_images = mnist.test_images()
test_images = (test_images / 255)
```

```python
# Prediction 10 = 0  
sample1 = test_images[278].reshape(784,)
sample2 = test_images[134].reshape(784,)
sample3 = test_images[567].reshape(784,)
sample4 = test_images[0].reshape(784,)
sample5 = test_images[978].reshape(784,)

pred1 =  nn.forward_prop(sample1)
pred2 =  nn.forward_prop(sample2)
pred3 =  nn.forward_prop(sample3)
pred4 =  nn.forward_prop(sample4)
pred5 =  nn.forward_prop(sample5)


print('Pred: ', np.argmax(pred1) + 1)
plt.imshow(sample1.reshape(28, 28))
plt.show()

print('Pred: ', np.argmax(pred2) + 1)
plt.imshow(sample2.reshape(28, 28))
plt.show()

print('Pred: ', np.argmax(pred3) + 1)
plt.imshow(sample3.reshape(28, 28))
plt.show()

print('Pred: ', np.argmax(pred4) + 1)
plt.imshow(sample4.reshape(28, 28))
plt.show()

print('Pred: ', np.argmax(pred5) + 1)
plt.imshow(sample5.reshape(28, 28))
plt.show()
```
